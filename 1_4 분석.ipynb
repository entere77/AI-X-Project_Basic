```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
seoul_business.columns
```

```python
# 사업체, 종사자 수 행과 합계,소계 행 제거
seoul_business=seoul_business.drop([0,1,2])
```

```python
seoul_business.head(10)
```

```python
seoul_business.info()
```

```python
# groupby 연산을 위한 처리

# 구 열은 임시로 저장
seoul_business_gu=seoul_business['자치구']
seoul_business_gu=seoul_business_gu.reset_index(drop=True)

# 구와 동 열을 제거
seoul_business=seoul_business.drop(['동','자치구'],axis=1)
```

```python
# '-'는 없는 것으로 판단, 0으로 치환
seoul_business=seoul_business.replace({'-':0})
```

```python
seoul_business.head()
```

```python
seoul_business_gu
```

```python
# 숫자 표기 시 따옴표로 인해 자료형 변경이 불가능
seoul_business.astype('int64')
```

```python
# 저장
seoul_business.to_csv("seoul_business.csv", index = False)
```

```python
# 읽어올 때 옵션으로 따옴표 제거
seoul_business=pd.read_csv('seoul_business.csv',sep=',',encoding='UTF-8',thousands=',')
```

```python
seoul_business.head(10)
```

```python
seoul_business.info()
```

```python
# 저장해둔 구 열과 병합
seoul_business1=pd.concat([seoul_business,seoul_business_gu],axis=1)
```

```python
seoul_business1.head()
```

```python
seoul_business1=seoul_business1.groupby('자치구').sum()
```

```python
seoul_business1['등록업종총합']=seoul_business1.sum(axis=1)
```

```python
seoul_business1.head()
```

```python
cor=seoul_business1.corr()
plt.figure(figsize=(20,10))
sns.heatmap(cor, annot=True)
plt.title('상관관계', size=30)
```

```python
# 구별 10개 등록업종의 사업체,종사자 수의 총 합
plt.figure(figsize=(20,10))
plt.plot('등록업종총합',data=seoul_business1)
```

```python
seoul_business1=seoul_business1.reset_index()
```

```python
seoul_business1.head()
```

```python
# 해당 데이터프레임을 'df_seoul_business.csv' 파일로 저장하세요.
seoul_business1.to_csv("df_seoul_business.csv", index = False)
```
