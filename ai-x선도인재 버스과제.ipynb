# 이제 주어진 자료로 정류장수가 부족한 구를 찾아봅니다.


## 첫번째로 구와 관련있는 상관계수를 찾아봅니다. heatmap 으로 나타냅니다.

```python


cor=df.corr()
plt.figure(figsize=(20,20))
sns.heatmap(cor, annot=True)
plt.title('상관관계', size=30)

```

### 정류장 변수와의 상관계수가 0.5보다 높게 나오는 변수에는 승하차총승객수와 65세이상고령자 변수가 존재했습니다.
### 또한, 상관계수가 0.3보다 높은 변수에는 부동산 중개 사업체수, 커피전문점, 기타주점업, 이동인구합 등의 변수들이 존재했습니다.

## 3.이변량 분석

* 자료의 종류에 맞게 X --> Y 에 대해서 그래프(시각화)와 가설검정(수치화)를 수행하고 결과를 평가합니다.

* 가설검정시 다음의 항목을 참조하여 수행합니다.
    * 유의수준 : 5%
    * 숫자 --> 숫자 : 상관분석
    * 범주 --> 범주 : 카이제곱검정
    * 범주 --> 숫자 : t검정, 분산분석
    * 숫자 --> 범주 : 로지스틱 회귀모형을 통해, 회귀계수의 P.value로 검정을 수행합니다.


### 가설1 이동인구수가 많을수록 더 많은 정류장 수가 있을 것이다.

```python
import scipy.stats as stats
```

```python
stats.pearsonr(df['이동인구(합)'],df['정류장수'])

# 양의 상관관계임을 알 수 있다.
# P > 0.05로 무의미
```

### 가설 2 승차총승객수가 많은 자치구에 정류장 수가 더 많을 것이다.

```python
stats.pearsonr(df['승차총승객수'],df['정류장수'])

# 양의 상관관계임을 알 수 있다.
# P < 0.05로 유의미
```

```python
ax = sns.lmplot(x='승차총승객수',y='정류장수', data=df)

# 가장 밑에 떨어진 데이터수가 중구입니다.
```

```python
plt.figure(figsize=(20,10))
plt.title('승차총승객수당 버스정류장수')
plt.bar(df['자치구'],sorted(df['정류장수']/df["승차총승객수"]))
plt.xlabel('구')
plt.ylabel('승차총승객수당 정류장수')


plt.show()
```

## 정류장수를 승차총승객수로 나눠서 승차총승객수 1인당 정류장수를 파악하였는데, 종로구와 중구가  낮게 나왔다.

### 가설 3 부동산 중개 및 대리업 사업체수가 많은 자치구에 정류장 수가 더 많을 것이다.

```python
stats.pearsonr(df['부동산 중개 및 대리업 사업체수'],df['정류장수'])

# 양의 상관관계임을 알 수 있다.
# P < 0.05로 유의미
```

```python
ax = sns.lmplot(x='부동산 중개 및 대리업 사업체수',y='정류장수', data=df)

# 가장 밑에 떨어져 있는 구는 중구이다.
```

```python
plt.figure(figsize=(20,10))
plt.title('부동산 중개 및 대리업 사업체')
plt.bar(df['자치구'],sorted(df['정류장수']/df["부동산 중개 및 대리업 사업체수"]))
plt.xlabel('구')
plt.ylabel('부동산 중개 사업체수 1개당 정류장수')


plt.show()
```

## 정류장수를 부동산 중개 및 대리업 사업체수로 나눠서 사업체수 1개당 정류장수를 파악하였는데, 종로구와 중구가 낮게 나왔다.
